# Importar Bibliotecas

```python
import pandas as pd
import os
import sqlite3
```

# Caminhos do Banco de Dados

```python
DB_PATH_ear_ons = os.path.join('../DADOS', 'banco_Monitor_Inteligencia_HidroFinanceira_SetorEletrico.db')

def executar_query(sql):
    with sqlite3.connect(DB_PATH_ear_ons) as conexao:
        return pd.read_sql(sql, conexao)
```

# Leitura dos arquivos

```python
df_ear_ons = executar_query("SELECT * FROM tb_ear_ons")

df_ear_ons.info()
```

```python
df_ear_ons.drop('id_subsistema',axis=1, inplace=True)

nomes_novos_ear = {
    'nom_subsistema':'estado',
    'ear_data':'data',
    'ear_max_subsistema':'ear_max',
    'ear_verif_subsistema_mwmes':'ear_mwmes',
    'ear_verif_subsistema_percentual':'ear_percentual',
}
df_ear_ons.rename(columns=nomes_novos_ear,inplace=True)

df_ear_ons
```
