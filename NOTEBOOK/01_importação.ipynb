# Importação de bibliotecas

```python
import pandas as pd
import os
import glob
```

# Caminhos base para arquivos

```python
caminho_silver = os.path.join('../DADOS/SILVER')


caminho_ons = os.path.join('../DADOS/RAW/ONS/EAR/EAR_DIARIO_SUBSISTEMA_*')
lista_arquivos_ear_ons = glob.glob(caminho_ons)
```

# Ajuste dos arquivos (Se necessário)

```python
# --- EAR: Consolidar todos os arquivos de anos, em um só arquivo "TOTAL" ---
lista_dfs_ear_ons = []

for arquivo in lista_arquivos_ear_ons:
    try:
        df_temp = pd.read_csv(arquivo, sep=';', decimal=',')
        lista_dfs_ear_ons.append(df_temp)
    except Exception as e:
        print(f"Erro ao ler {arquivo}: {e}")

if lista_dfs_ear_ons:
    df_ear_ons = pd.concat(lista_dfs_ear_ons, ignore_index=True)
```

# Leitura de Dados & Salvar Arquivos

```python
df_ear_ons.to_csv(os.path.join(caminho_silver,'EAR_DIARIO_SUBSISTEMA_TOTAL'))
```
