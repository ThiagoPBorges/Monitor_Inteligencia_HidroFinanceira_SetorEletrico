# Importação de bibliotecas

```python
import pandas as pd
import os
import glob
import sqlite3
```

# Caminhos base para arquivos

```python
caminho_dados = os.path.join('../DADOS')
caminho_banco = os.path.join(caminho_dados, 'banco_Monitor_Inteligencia_HidroFinanceira_SetorEletrico.db')
conexao = sqlite3.connect(caminho_banco)

caminho_ons = os.path.join('../DADOS/RAW/ONS/EAR/EAR_DIARIO_SUBSISTEMA_*')
lista_arquivos_ear_ons = glob.glob(caminho_ons)
```

# Ajuste dos arquivos (Se necessário)

```python
# --- EAR: Consolidar todos os arquivos de anos, em um só arquivo "TOTAL" ---
lista_dfs_ear_ons = []

for arquivo in lista_arquivos_ear_ons:
    try:
        df_temp = pd.read_csv(arquivo, sep=';', decimal=',')
        lista_dfs_ear_ons.append(df_temp)
    except Exception as e:
        print(f"Erro ao ler {arquivo}: {e}")

if lista_dfs_ear_ons:
    df_ear_ons = pd.concat(lista_dfs_ear_ons, ignore_index=True)
```

# Leitura de Dados & Salvar no Banco de Dados

```python
# Enviar o df_ear_ons para o banco de dados
df_ear_ons.to_sql(
    name='tb_ear_ons',
    con=conexao,
    if_exists='replace',
    index=False
)

# Fechar conexão
conexao.close()

print(f"df_ear_ons salvo com sucesso no banco: {caminho_banco}")
```
